### Steps to Check Data Balance
To check if our dataset is balanced, we can analyze the distribution of the target variable (e.g., the y column or class labels)


**1. Check the Value Counts of the Target Variable: Use the .value_counts() function to see the number of samples in each class.**

**2. Calculate the Percentage of Each Class: Divide the count of each class by the total number of samples.**

**As class is almost balanced we don't require any resampling Techniques**

### Additional Tips for Handling Imbalance
If our dataset is imbalanced, we can consider using techniques such as:

1. Resampling:

- Oversampling the minority class (e.g., using SMOTE).
- Undersampling the majority class.

2. Using Class Weights:

Set class_weight='balanced' in your classifier (e.g., LogisticRegression, RandomForest).

3. Choose Metrics for Imbalanced Data:

Use metrics like precision, recall, F1-score, or ROC-AUC.
These steps helps ensuring our model performs well even with imbalanced data.

**When checking for balance, we are primarily concerned with the target variable (dependent variable).** 

**This is because:**

**Why Focus on the Target Variable?**

 1. Target Variable Represents the Prediction Goal:
In supervised learning, the goal is to predict the target variable. If the target classes are imbalanced, the model may become biased towards the majority class, leading to poor performance for the minority class.

 2. Impact on Model Training:
Machine learning algorithms typically assume that the classes in the target variable are equally represented. An imbalanced target variable can lead to biased predictions where the model favors the majority class.

 3. Features Don't Require Balance:

- Feature variables (independent variables) don’t need to be balanced. Their purpose is to provide diverse and informative input for predicting the target. However, we do check for data quality issues like:
    - Missing values.
    - Outliers.
    - Skewed distributions (e.g., highly imbalanced numeric features).

- Feature engineering and preprocessing handle these issues separately.


**What About Features/Independent Variables?**
Although balance isn't required for features, we may still want to check for:

1. Skewness in Numeric Features:
Skewed features can affect certain models like Logistic Regression or SVM. Transformations like log-scaling or normalization can help.


```python
# Checking skewness
skewness = data.skew()
print(skewness)
```

**General Interpretation of Skewness**

- Skewness = 0: The distribution is perfectly symmetrical (normal distribution).
- Skewness > 0: The distribution is positively skewed (longer tail on the right side).
- Skewness < 0: The distribution is negatively skewed (longer tail on the left side).

**Our Skewness Values**

1. Features with Skewness ≈ 0:

    - Example: class (0.071946), gill-color (0.061410)
    - These features are approximately symmetric and don't show significant skewness.

2. Features with Positive Skewness:

    - Examples: cap-color (0.706965), gill-spacing (1.840088), stalk-root (0.947852)
    - The data is positively skewed, meaning the distribution has a longer right tail.
    - A majority of the data is concentrated on the lower values.

3. Features with Negative Skewness:

    - Examples: cap-surface (-0.590859), gill-attachment (-5.977076), stalk-surface-above-ring (-1.098739)
    - The data is negatively skewed, meaning the distribution has a longer left tail.
    - A majority of the data is concentrated on the higher values.

4. Features with Extreme Skewness:

    - Example: veil-color (-6.946944), gill-attachment (-5.977076)
    - These features are heavily skewed. Extreme skewness can affect machine learning models and may require transformation (e.g., log, square root) to normalize the data.

5. Feature with Zero Skewness:

    - Example: veil-type (0.000000)
    - This feature has no skewness, indicating it is perfectly symmetric. In categorical data, this might indicate an equal or very balanced distribution.

### 9 . PCA 

**Dimensionality Reduction Technqiue**
**(we can remove highly correlated columns/independent variables so that dataset becomes smaller in size and still hold  most of information)**

```python
from sklearn.decomposition import PCA
pca1 = PCA(n_components = 7) # n gives no. of components which we want , if we want to retain 90% info we can give n= 0.90
pca_fit = pca1.fit_transform(X) # independent variables after applying pca
```

```python
pca1.explained_variance_ratio_
```

**As 23 features it produced new features 7 components**

```python
sum(pca1.explained_variance_ratio_) # 7 components retaining 85% info from dataset
```

### 10. Split dataset into Training and Test Set

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(pca_fit, y ,test_size=0.20,random_state=42)
```

### 11. Import Models

```python
data.head()
```

**As target variable 'class' is categorical variable hence well use classification models**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

### 12. Model Training

```python
lr = LogisticRegression()
lr.fit(X_train,y_train)

knn = KNeighborsClassifier()
knn.fit(X_train,y_train)

svc = SVC()
svc.fit(X_train,y_train)

dt = DecisionTreeClassifier()
dt.fit(X_train,y_train)

rm = RandomForestClassifier()
rm.fit(X_train,y_train)

gb = GradientBoostingClassifier()
gb.fit(X_train,y_train)
```

### 13. Prediction on Test Data(X_test) using unseen samples

```python
y_pred1 = lr.predict(X_test)
y_pred2 = knn.predict(X_test)
y_pred3 = svc.predict(X_test)
y_pred4 = dt.predict(X_test)
y_pred5 = rm.predict(X_test)
y_pred6 = gb.predict(X_test)
```

### 14. Evaluating the Algorithm

```python
from sklearn.metrics import accuracy_score
```

```python
print("Accuracy of LR :",accuracy_score(y_test,y_pred1)) # actual values stored in y_test
print("Accuracy of KNN :",accuracy_score(y_test,y_pred2))
print("Accuracy of SVC :",accuracy_score(y_test,y_pred3))
print("Accuracy of DT :",accuracy_score(y_test,y_pred4))
print("Accuracy of RM :",accuracy_score(y_test,y_pred5))
print("Accuracy of GB :",accuracy_score(y_test,y_pred6))
```

In this case, accuracy is being used as the evaluation metric because the problem likely involves binary classification (e.g., edible vs. poisonous mushrooms) and the dataset might be balanced (i.e., roughly equal numbers of edible and poisonous mushrooms)

```python
final_data = pd.DataFrame({'Models': ['LR','KNN','SVC','DT','RM','GB'],
             'ACC' :[accuracy_score(y_test,y_pred1)*100,
                     accuracy_score(y_test,y_pred2)*100,
                     accuracy_score(y_test,y_pred3)*100,
                     accuracy_score(y_test,y_pred4)*100,
                     accuracy_score(y_test,y_pred5)*100,
                     accuracy_score(y_test,y_pred6)*100,
                    ]})
```
